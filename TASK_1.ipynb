```python
# IMPORTING LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({'Age':[25,30,28,32,22,31,35,29,49,26,30,49,57,34,55,26,37,47,42,45,47,45,52,65,37,27,52,50,
        36,29,62,67,66,67,60,35,22,21,42,23,27,29,31,30,32,35,23,39,40,56,65,60,70,15],
        'Gender':['Male','Female','Female','Male','Female','Male','Female',
        'Female','Male','Female','Male','Female','Female','Male','Female','Male','Female','Female','Male','Female','Male','Female','Female','Male','Male',
        'Male','Female','Female','Male','Female','Male','Female','Female','Male','Female','Male','Female','Female','Male','Female','Male','Female','Female','Male','Male','Male','Female','Female','Male','Male'
        ,'Male','Female','Female','Male']})


        
```

```python
# GROUPING DATA TOGETHER AND COUNTING OCCURENCES

Age_Gender_counts = data.groupby(['Age','Gender']).size().unstack(fill_value=0)
```

```python
#DEFINING COLORS FOR BARS
colors = ['Blue','Pink']


# PLOTTING BAR CHART
ay = Age_Gender_counts.plot(kind='bar',figsize=(10,6),color=colors)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Age and Gender")
plt.legend(title='Gender',labels = Age_Gender_counts.columns)
plt.show()
```
